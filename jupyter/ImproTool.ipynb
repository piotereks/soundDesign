```python

import pandas as pd
import pprint
from collections import OrderedDict
from datetime import datetime


import time
import mido

fname='New Improvisation tool + chords_mod_For_YNB.xlsx'
 

scales_df=None
rythmics_df=None
phrases_df=None

def load_xlsx():
    
    global scales_df
    global rythmics_df
    global phrases_df
    # only read specific columns from an excel file
    require_cols = ['Type','Nazwa skali','Scale_len','Tony']
    scales_df = pd.read_excel(fname, \
                              sheet_name='Skale', usecols = require_cols) 
    scales_df=scales_df[['Type','Nazwa skali','Scale_len','Tony']]
#     display(scales_df)
    # rythmics_df = pd.read_excel(fname, \
    #                           sheet_name='Rytmika',header= 1, 
    #                             usecols = [cols for cols in range(69)]).fillna('')
    phrases_df = pd.read_excel(fname, sheet_name='Ogrywanie skal',header= 1, usecols = [cols for cols in range(69)]).fillna('')

def phrases():
    def pcs_phrases():
        global phrases_dfx
        start_col_idx = phrases_df.columns.get_loc('Step vs P1')
        phrases_df.insert(loc = start_col_idx+1,column = 'Pattern',value=None)
        pattern_col_idx = phrases_df.columns.get_loc('Pattern')
#         display(phrases_df)
        # phrases_dfx=phrases_df.head(5).copy()
        phrases_dfx=phrases_df.copy()


        # display(phrases_dfx)
        # for x,y in xsub_df.iterrows():
        #   print(x,y)

#         print(start_col_idx)
        for i in range(phrases_dfx.shape[0]):
            clearFlag=False
            for j in range(start_col_idx,phrases_dfx.shape[1]):
                if phrases_dfx.iloc[i, j]=='x' or clearFlag:
                    phrases_dfx.iloc[i, j]=None
                    clearFlag=True

            pattern=[interval for interval in phrases_dfx.iloc[i, start_col_idx+1:] if interval!=None]
#             print(pattern)
            phrases_dfx.iat[i,pattern_col_idx]=pattern

    def write_wrk_phrases():
        # Getting the current date and time
        dt = datetime.now().strftime("_%Y%m%d_%H%M%S")
        filename="wrk_phrases_dict.xlsx"
        try:
            with pd.ExcelWriter(filename) as writer:
                phrases_dfx.to_excel(writer, sheet_name="Phrases", index=False)
        except:
            filename=filename.split('.')[0]+ dt+'.xlsx'
            with pd.ExcelWriter(filename) as writer:
                phrases_dfx.to_excel(writer, sheet_name="Phrases", index=False)                        
            
    pcs_phrases()

    display(phrases_dfx)
#     print(1)
    # pprint.pprint(phrases_dfx.loc[:,:'Pattern'].to_dict(orient="index"))
#     print(2)
    # preferred
#     phrases_dict=phrases_dfx.loc[:,:'Pattern'].to_dict(orient="records")
    # pprint.pprint(phrases_dict)
    write_wrk_phrases()

def scales():
    def pcs_scales():
        global scales_dfx
        scales_dfx=scales_df.copy()
#         print('blah')
        tones_list_col_idx = scales_dfx.columns.get_loc('Tony')
        scales_dfx.insert(loc=tones_list_col_idx,column = 'Tones_list',value=None)

        for i in range(scales_dfx.shape[0]):
            scales_dfx.iat[i,tones_list_col_idx]=[scale_step for scale_step in scales_dfx.iloc[i, tones_list_col_idx+1].split(' ') \
                          if scale_step!='']

        display(scales_dfx)
    pcs_scales()
```

```python

    
# https://www.twilio.com/blog/working-with-midi-data-in-python-using-mido    
# https://mido.readthedocs.io/en/latest/

def main():


    load_xlsx()
    phrases()
    scales()


        
if __name__== "__main__":
    main()
    print("Processing Done")
```
